```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver import ActionChains
import mysql.connector 

df_ap = pd.read_csv('df_AP.csv')
df_ap
```

```python
# AP(ANDRA PRADESH)

driver = webdriver.Chrome()
driver.maximize_window()
apsrtc_bus_name = []
apsrtc_bus_type = []
apsrtc_departing_time = []
apsrtc_duration = []
apsrtc_reaching_time = []
apsrtc_star_rating = []
apsrtc_price = []
apsrtc_route_names = []
apsrtc_route_link = []
apsrtc_seat = []
for i , r in df_ap.iterrows():
    link = r['APSRTC_links'] # ,
    routes = r['APSRTC_routes']

    driver.get(link)
    time.sleep(5)

    elements = driver.find_elements(By.XPATH,f"//a[contains(@href,'{link}')]")
    for url in elements:
        url.click()
        time.sleep(3)
    try:
        buttons = driver.find_element(By.XPATH,"//div[@class = 'button']")
        buttons.click()

    except:
        pass
    time.sleep(4)    


    previous_height = 0
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Wait for content to load
        new_height = driver.execute_script("return document.body.scrollHeight;")
        if new_height == previous_height:
            break  # Stop when no new content is loaded
        previous_height = new_height
    
      

    bus_box = driver.find_elements(By.XPATH,'//div[@class="clearfix row-one"]')
    if not bus_box:
               pass
    else: 
        for busitems in bus_box:

            try:#travels lh-24 f-bold d-color
                Bus_name = busitems.find_element(By.XPATH,".//div[@class ='travels lh-24 f-bold d-color']")   
            except:
                Bus_name = None
            try:
                Bus_type = busitems.find_element(By.XPATH,'.//div[@class="bus-type f-12 m-top-16 l-color evBus"]')
            except:
                Bus_type = None 
            try: #dp-time f-19 d-color f-bold
                Departing_time = busitems.find_element(By.XPATH,".//div[@class='dp-time f-19 d-color f-bold']")
            except:
                    Departing_time = None
            try:        #dur l-color lh-24
                Duration = busitems.find_element(By.XPATH,'.//div[@class="dur l-color lh-24"]')
            except:
                    Duration = None
            try:   #bp-time f-19 d-color disp-Inline
                Reaching_time = busitems.find_element(By.XPATH,'.//div[@class="bp-time f-19 d-color disp-Inline"]')
            except:
                Reaching_time = None
            try:    
                price = busitems.find_element(By.XPATH,'.//div[@class="fare d-block"]')
            except:
                price = '0' 

            try:
                star_rating = busitems.find_element(By.XPATH, './/div[@class = "rating-sec lh-24"]').text
                star_rating = star_rating if star_rating.strip() else '0.0'
            except:
                 star_rating = '0.0'
                 
            try: # seat 
                seat = busitems.find_element(By.XPATH,'.//*[contains(@class, "seat-left")]').text
                seat = seat if seat.strip() else 'no seats available'
            except:
                 seat = 'no seats available'

        # Append details to respective lists
            apsrtc_bus_name.append(Bus_name.text)
            apsrtc_bus_type.append(Bus_type.text)
            apsrtc_departing_time.append(Departing_time.text)
            apsrtc_duration.append(Duration.text)
            apsrtc_reaching_time.append(Reaching_time.text)
            apsrtc_price.append(price.text)
            apsrtc_star_rating.append(star_rating)
            apsrtc_route_link.append(link)
            apsrtc_route_names.append(routes)    
            apsrtc_seat.append(seat)    

        
```

```python
insert_sql = """insert into all_bus_details(route_name,
                                       route_link,
                                       busname,
                                       bustype,
                                       departing_time,
                                       duration,
                                       reaching_time,
                                       star_rating,
                                       price,
                                       seats_available,
                                       state_name)
                                    Values(%s, %s, %s, %s, %s , %s, %s, %s, %s, %s,%s)"""
  # Convert DataFrame to a list of tuples
values = [tuple(row) for row in df_read.values]
try:
    cursor = connection.cursor()
    batch_size = 1000
    for i in range(0, len(values), batch_size):
        batch_values = values[i:i+batch_size]
        cursor.executemany(insert_sql, batch_values)
        connection.commit()  # Commit the batch after insertion

    print("Data inserted successfully!")

except Exception as e:
    print(f"Error occurred: {e}")
    
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from selenium.webdriver import ActionChains
import mysql.connector 
from selenium.webdriver.common.action_chains import ActionChains


try: 
        driver = webdriver.Chrome()
        
        opening = driver.get('https://www.redbus.in/');
        print('url opened successfully')
        time.sleep(5)
                            
except:
            
        print('Error occur when open')

def maximize_window():
    try:
        driver.maximize_window()
        print('window maximized')
    
    except:
        print('Error when maximizing window')
  
maximize_window()    
APSRTC_link_list = []
APSRTC_route_list = []
def APSRTC_bus_details():
  
    driver.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")

    
    
    try:
        
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "#root > div > div.D117_main.D117_container"))
        )

        # Locate the pagination container
        pagination_container = driver.find_element(By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')

        # Get all page elements within the container
        pages = pagination_container.find_elements(By.XPATH, './div') 

        for page in pages:
            try:
            
                driver.execute_script("arguments[0].scrollIntoView(true);", page)

                WebDriverWait(driver, 10).until(EC.element_to_be_clickable(page))

                ActionChains(driver).move_to_element(page).click().perform()

                WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, "#root > div > div.D117_main.D117_container"))
                )

               
                items = driver.find_elements(By.XPATH,"//a[@class='route']")
                for item in items:
                    route_name = item.text
                    link = item.get_attribute('href')
                    APSRTC_link_list.append(link)
                    APSRTC_route_list.append(route_name)

            except Exception as e:
                print(f"Error processing page: {e}")

    except Exception as e:
        print(f"An error occurred during pagination: {e}")

    
    # driver.quit()
    

APSRTC_bus_details()

df = pd.DataFrame({'APSRTC_links':APSRTC_link_list,
             'APSRTC_routes':APSRTC_route_list})


df
```
